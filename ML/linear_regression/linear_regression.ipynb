```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
df
```

```python
from sklearn import linear_model
```

```python
plt.xlabel('Year')
plt.ylabel('per capita income(US$)')
plt.title('capita distribution over the years')
plt.scatter(df[['year']], df['per capita income (US$)'], marker='+')
```

```python
capita = df['per capita income (US$)']
```

```python
capita.head()
```

```python
year = df.drop('per capita income (US$)', axis='columns')
```

```python
year.head()
```

```python
reg = linear_model.LinearRegression()
reg.fit(year, capita)
```

```python
prediction_data = pd.DataFrame({'year': [2020, 2021]})
reg.predict(prediction_data)
```

```python
reg.coef_
```

```python
reg.intercept_
```
